```python
data = 9435

row = [0] * 300
matrix = [row] * 300

print(matrix[0])
```

```python
def compute_score(x, y, test_input = data):
    rack_id = x + 10
    power_level = rack_id * y
    power_level += test_input
    product = power_level * rack_id
    
    if product < 100:
        return -5

    product = str(product)
    digit = product[-3]
    return int(digit) - 5

print(compute_score(122, 79, 57))
print(compute_score(217, 196, 39))
print(compute_score(101, 153, 71))
```

```python
for y in range(300):
    for x in range(300):
        matrix[y][x] = compute_score(x, y)
```

```python
best_power = 0
best_pair = []
for x in range(300 - 3):
    for y in range(300 - 3):
        sum_for_index = 0
        for x_check in range(3):
            for y_check in range(3):
                sum_for_index += compute_score(x + x_check, y + y_check)
        
        if sum_for_index > best_power:
            best_power = sum_for_index
            best_pair = [x, y]
```

```python
print(best_power)
print(best_pair)
```

```python
# Part 2
best_power = 0
best_pair = []
size = 0
for x in range(300 - 3):
    for y in range(300 - 3):
        sum_for_index = 0
        
        for square_value in range(min(20, 300 - max(x, y))):
            for x_check in range(square_value):
                for y_check in range(square_value):
                    sum_for_index += compute_score(x + x_check, y + y_check)

            if sum_for_index > best_power:
                best_power = sum_for_index
                best_pair = [x, y]
                size = square_value
                
    print("Done with row " + str(x))
```

```python
print(best_power)
print(best_pair)
```
